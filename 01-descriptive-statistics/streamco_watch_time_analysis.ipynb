StreamCo Watch Time Analysis - exploring central tendency on a small dataset to learn when mean vs. median is appropriate.

Marketing wants to know what 'typical user engagement' looks like for a press release.

```python
import statistics

watch_time = [12, 18, 22, 25, 28, 30, 31, 33, 35, 36,
              38, 40, 42, 45, 47, 48, 50, 52, 55, 58,
              62, 65, 70, 75, 82, 95, 110, 145, 220, 380]

print(f"Count: {len(watch_time)} users")
print(f"Mean: {statistics.mean(watch_time):.1f} minutes")
print(f"Median: {statistics.median(watch_time):.1f} minutes")
print(f"Std Dev: {statistics.stdev(watch_time):.1f} minutes")
print(f"Min: {min(watch_time)} minutes")
print(f"Max: {max(watch_time)} minutes")

q1 = statistics.quantiles(watch_time, n=4)[0]
q3 = statistics.quantiles(watch_time, n=4)[2]
print(f"Q1 (25th percentile): {q1:.1f} minutes")
print(f"Q3 (75th percentile): {q3:.1f} minutes")
print(f"IQR: {q3 - q1:.1f} minutes")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(watch_time, bins=15, edgecolor='black', color='#378ADD')
plt.axvline(statistics.mean(watch_time), color='red', linestyle='--', label=f'Mean: {statistics.mean(watch_time):.0f}')
plt.axvline(statistics.median(watch_time), color='green', linestyle='--', label=f'Median: {statistics.median(watch_time):.0f}')
plt.xlabel('Watch time (minutes)')
plt.ylabel('Number of users')
plt.title('Daily watch time distribution at StreamCo')
plt.legend()
plt.show()
```

The typical StreamCo user watches about 48 minutes per day (median), with most users falling between 33 and 71 minutes (the middle 50%). The mean of 68 minutes is misleading here because the distribution is right-skewed - a small number of heavy users (145-380 minutes/day) inflate the average. These are likely binge-watchers or accounts left running, and they don't reflect typical behavior. Recommendation: use the median for the press release; "48 minutes per day" is honest and representative.

I learned that mean and median tell different stories when data is skewed. I learned to always plot before reporting.
